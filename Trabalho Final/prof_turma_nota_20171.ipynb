```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
%matplotlib inline
```

```python
filez = pd.read_csv('prof_turma_nota_imd_20171.csv', sep=';', error_bad_lines=False)

filez_prof_col = filez['professor']
filez_turma_col = filez['aula']
filez_aluno_col = filez['aluno']
filez_media_col = filez['media']


myDic = {}

valores = {}

prev = ""
now = ""
val =0
iterator=0
for prof,aluno,media,turma in zip(filez_prof_col,filez_aluno_col,filez_media_col,filez_turma_col):
	concatena = turma+" "+prof
	#now = turma
	now = concatena
	if( prev != now):
		valores[concatena] = 1
		myDic[concatena] = media
	else:
		valores[concatena] += 1
		myDic[concatena] += media
        
	prev = now
	
vetx = []
vety = []


veti = [len(myDic)]
iterator = 0
for i,j,z in zip(myDic.values(),valores.values(),myDic):
	veti[iterator] = iterator
	vetx.append( float(i/j))
	vety.append(z)

ax = plt.axes()
ax.set_xlabel("materia + professor",fontsize=25)
ax.set_ylabel("media de notas da turma",fontsize=25)

fig = plt.gcf()
fig.set_size_inches(20,5)

#plt.bar(range(len(myDic)),myDic.values(),align="center",width=0.5)
#plt.xticks(range(len(myDic)),myDic.keys(),rotation="vertical",fontsize=10)
plt.title("Grafico de materias de 2017.1",fontsize=25)
plt.bar(range(len(vetx)),vetx,align="center",width=0.5)
plt.xticks(range(len(myDic)),vety,rotation="vertical",fontsize=15)
```
