```python
import pandas as pd
from bokeh.io import output_notebook, show
from bokeh.layouts import widgetbox
from bokeh.models.widgets import Button, RadioButtonGroup, Select, Slider
```

```python
pd_dados = pd.read_csv('turma_professor_notas.csv')
pd_disciplinas = pd.read_csv('disciplinas.csv')
```

```python
def my_text_input_handler(attr, old, new):
    print("Previous label: " + old)
    print("Updated label: " + new)

def my_radio_handler(new):
    print ('Radio button option ' + str(new) + ' selected.')
```

```python
output_notebook()
selectProf = Select(title="Professores:", value="foo", options=pd_dados['nome_prof'].unique().tolist())
selectDisc = Select(title="Disciplinas:", value="foo", options=pd_dados['nome_disc'].unique().tolist())

selectProf.on_change("value", my_text_input_handler)
#selectProf.on_click(my_radio_handler)

# put the results in a row
show(widgetbox(selectProf,selectDisc, width=300))
```
