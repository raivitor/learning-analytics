Neste notebook será mostrado a "eficiencia" de um professor. Fiz uma alteração no csv dos alunos e agora tem a quantidade de reprovações na mesma disciplina. Então eu peguei as disciplinas C1 e FMC1 e vi quem tinha passado de primeira. Com esses dados em mão eu fui ver quem era os professores deles em RPMTI e depois disso a porcentagem de aprovação.<br>
Com FMC2 eu fiz bem parecido só que ao invés de comparar com RPMTI eu comparei com FMC1.

```python
import pandas as pd
pd.options.display.float_format = '{:,.2f} %'.format
%matplotlib inline
```

```python
pd_dados = pd.read_csv('turma_professor_notas.csv')
pd_disciplinas = pd.read_csv('disciplinas.csv')
pd_alunos = pd.read_csv('turma_aluno2.csv')
```

```python
def compDisci(disc1, disc2):
    pd_c1 = pd_alunos[pd_alunos['nome_disc'] == disc1]
    pd_c1_1 = pd_c1[pd_c1['tentativas_disc'] == 1]

    listProf = []
    for index, row in pd_alunos[pd_alunos['nome_disc'] == disc2].iterrows():
        if(not(pd_c1_1[pd_c1_1['discente'] == row['discente']]).empty):
            if(row['nome_prof'] != 'vazio'):
                listProf.append(row['nome_prof'])
                
    return pd.Series(listProf) 

def alunosDisc(disc):
    listProf = []
    for index, row in pd_alunos[pd_alunos['nome_disc'] == disc].iterrows():
        if(row['nome_prof'] != 'vazio'):
            listProf.append(row['nome_prof'])
    return pd.Series(listProf) 
```

## Comparando professores que ensinaram RPMTI e seus alunos que passaram em C1 de primeira

```python
pd_prof = compDisci("CÁLCULO DIFERENCIAL E INTEGRAL I", "RESOLUÇÃO DE PROBLEMAS MATEMÁTICOS PARA TI")
totalAlunos = alunosDisc("RESOLUÇÃO DE PROBLEMAS MATEMÁTICOS PARA TI")
print(((pd_prof.value_counts()/totalAlunos.value_counts())*100).sort_values(ascending = False))
((pd_prof.value_counts()/totalAlunos.value_counts())*100).sort_values(ascending = False).plot(kind='bar', figsize=(14, 6))
```

## Comparando professores que ensinaram RPMTI e seus alunos que passaram em FMC1 de primeira

```python
pd_prof = compDisci("FUNDAMENTOS MATEMÁTICOS DA COMPUTAÇÃO I", "RESOLUÇÃO DE PROBLEMAS MATEMÁTICOS PARA TI")
totalAlunos = alunosDisc("RESOLUÇÃO DE PROBLEMAS MATEMÁTICOS PARA TI")
print(((pd_prof.value_counts()/totalAlunos.value_counts())*100).sort_values(ascending = False))
((pd_prof.value_counts()/totalAlunos.value_counts())*100).sort_values(ascending = False).plot(kind='bar', figsize=(14, 6))
```

## Comparando professores que ensinaram FMC1 e seus alunos que passaram em FMC2 de primeira

```python
pd_prof = compDisci("FUNDAMENTOS MATEMÁTICOS DA COMPUTAÇÃO II", "FUNDAMENTOS MATEMÁTICOS DA COMPUTAÇÃO I")
totalAlunos = alunosDisc("FUNDAMENTOS MATEMÁTICOS DA COMPUTAÇÃO I")
print(((pd_prof.value_counts()/totalAlunos.value_counts())*100).sort_values(ascending = False))
((pd_prof.value_counts()/totalAlunos.value_counts())*100).sort_values(ascending = False).plot(kind='bar', figsize=(14, 6))
```
